# Ktav Ivri Genesis

This document introduces the usage of the [Gematria](Gematria wiki) on the book of Genesis. There are several number mappings you can use in Gematria. Here I have implemented Ktav Ivri Hebrew letter number substitutes. See: The [Gematria of Ktav Ivri](https://www.academia.edu/23773185/The_Gematria_of_Ktav_Ivri) by BethSheba Ashe for a reference.

This document is created with Python and Javascript so that underlying text manuscript ([Modern Hebrew Bible](http://unbound.biola.edu/index.cfm?method=downloads.showDownloadMain)) is loaded, then indexed by Hebrew characters so that at the end it is possible to find gematria values from the book.
